# Model_Loss
#### Compare the loss/val_loss between models in a directory

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import glob
```

```python
def plot_loss(pathname, slice_left=0, **kwargs):
    """
    Plot the loss from model training
    
    Arguments:
        pathname: path to history file(s)
        slice_left: left index of slice to plot
    
    Returns:
        plots
    
    """
    paths = glob.glob(pathname+'history*');
    losses = [pd.read_csv(path) for path in paths];
    
    n = len(losses);
    pn = len(pathname);
    
    print('Number of history files found:', n, '\n')
    for loss, path in zip(losses, paths):
        print(path[pn:-4], '\n', loss.min(), '\n')
    
    fig, axs = plt.subplots(1,n, figsize=(10,10))

    for ax, i in zip(axs.ravel(), np.arange(n)):
        plot(ax, losses[i], slice_left, **kwargs)
        ax.set_title(paths[i][pn:])
    

def plot(ax, loss, slice_left, **kwargs):
    ax.plot(loss.loss[slice_left:], **kwargs)
    ax.plot(loss.val_loss[slice_left:], **kwargs)
#    plt.yscale('log')
    ax.legend()
```

# 3Layers: (64,32,16), epochs: 4k, act: relu

```python
plot_loss('hyalite/3layers/', slice_left=1, alpha=0.8)
```

# 2Layers: (30, 10), epochs: 4k

```python
plot_loss('hyalite/2layers/', slice_left=50, alpha=0.8)
```
